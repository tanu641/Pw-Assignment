```python
# 1MB= 1024KB
def size_process(item):
    if str(item)[-1]== "M":
        result=float(str(item).replace("M", ""))
        result= result*1024
        return result
    
    elif str(item)[-1]== "K":
        result=float(str(item).replace("K", ""))
        return result
    else:
        return str(np.nan)
    

```

```python
size_process("19M")
```

```python
df_copy['Size']= df_copy['Size'].apply(size_process)
```
