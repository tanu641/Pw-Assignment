```python
#1. When to Use For Loop and While Loop (with Examples):
For Loop:

Use when you know the exact number of iterations or are iterating over a sequence (e.g., list, range, string).
Example:


for i in range(5):
    print(i)  # Prints numbers from 0 to 4
While Loop:

Use when the number of iterations is not predetermined and depends on a condition.
Example:


count = 0
while count < 5:
    print(count)
    count += 1  # Prints numbers from 0 to 4


#Q2. Program to Print the Sum and Product of First 10 Natural Numbers


# Using for loop
sum_for = 0
product_for = 1
for i in range(1, 11):
    sum_for += i
    product_for *= i
print(f"Using for loop: Sum = {sum_for}, Product = {product_for}")

# Using while loop
sum_while = 0
product_while = 1
num = 1
while num <= 10:
    sum_while += num
    product_while *= num
    num += 1
print(f"Using while loop: Sum = {sum_while}, Product = {product_while}")
Q3. Program to Compute the Electricity Bill for a Household


# Input from user
units = int(input("Enter the electricity units consumed: "))

# Calculate bill
if units <= 100:
    bill = units * 4.5
elif units <= 200:
    bill = (100 * 4.5) + (units - 100) * 6
elif units <= 300:
    bill = (100 * 4.5) + (100 * 6) + (units - 200) * 10
else:
    bill = (100 * 4.5) + (100 * 6) + (100 * 10) + (units - 300) * 20

print(f"The total electricity bill is: Rs. {bill}")
Q4. Program to Compute Cubes and Filter Numbers Divisible by 4 or 5

# List of numbers from 1 to 100
numbers = list(range(1, 101))
filtered_numbers = []

# Using for loop
for num in numbers:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        filtered_numbers.append(num)
print(f"Using for loop: {filtered_numbers}")

# Using while loop
filtered_numbers_while = []
i = 1
while i <= 100:
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        filtered_numbers_while.append(i)
    i += 1
print(f"Using while loop: {filtered_numbers_while}")



Q5. Program to Count Vowels in a String

string = "I want to become a data scientist"
vowels = "aeiouAEIOU"
vowel_count = 0

for char in string:
    if char in vowels:
        vowel_count += 1
















```
